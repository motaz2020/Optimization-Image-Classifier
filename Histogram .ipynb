```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the RGB image
image = cv2.imread('/content/IMG_6973.JPG')

# Convert the image to HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Separate the channels
h, s, v = cv2.split(hsv_image)

# Apply histogram equalization on the V channel
equalized_v = cv2.equalizeHist(v)

# Merge the equalized V channel with the original H and S channels
equalized_hsv = cv2.merge([h, s, equalized_v])

# Convert the image back to the RGB color space
equalized_image = cv2.cvtColor(equalized_hsv, cv2.COLOR_HSV2BGR)

# Plot the histograms of the three channels after equalization
plt.figure(figsize=(10, 5))

# Original image histograms
plt.subplot(2, 2, 1)
plt.hist(image[:, :, 0].ravel(), bins=256, color='red', alpha=0.5, label='Red')
plt.hist(image[:, :, 1].ravel(), bins=256, color='green', alpha=0.5, label='Green')
plt.hist(image[:, :, 2].ravel(), bins=256, color='blue', alpha=0.5, label='Blue')
plt.title('Original Image Histograms')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.legend()

# Equalized image histograms
plt.subplot(2, 2, 2)
plt.hist(equalized_image[:, :, 0].ravel(), bins=256, color='red', alpha=0.5, label='Red')
plt.hist(equalized_image[:, :, 1].ravel(), bins=256, color='green', alpha=0.5, label='Green')
plt.hist(equalized_image[:, :, 2].ravel(), bins=256, color='blue', alpha=0.5, label='Blue')
plt.title('Equalized Image Histograms')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.legend()

# Show the images and histograms
plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB))
plt.title('Equalized Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
